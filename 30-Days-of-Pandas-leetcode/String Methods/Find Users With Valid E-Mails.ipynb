# Find Users With Valid E-Mails


```python
Table: Users

+---------------+---------+
| Column Name   | Type    |
+---------------+---------+
| user_id       | int     |
| name          | varchar |
| mail          | varchar |
+---------------+---------+
user_id is the primary key (column with unique values) for this table.
This table contains information of the users signed up in a website. Some e-mails are invalid.
 

Write a solution to find the users who have valid emails.

A valid e-mail has a prefix name and a domain where:

The prefix name is a string that may contain letters (upper or lower case), digits, underscore '_',
period '.', and/or dash '-'. The prefix name must start with a letter.
The domain is '@leetcode.com'.
Return the result table in any order.

The result format is in the following example.

 

Example 1:

Input: 
Users table:
+---------+-----------+-------------------------+
| user_id | name      | mail                    |
+---------+-----------+-------------------------+
| 1       | Winston   | winston@leetcode.com    |
| 2       | Jonathan  | jonathanisgreat         |
| 3       | Annabelle | bella-@leetcode.com     |
| 4       | Sally     | sally.come@leetcode.com |
| 5       | Marwan    | quarz#2020@leetcode.com |
| 6       | David     | david69@gmail.com       |
| 7       | Shapiro   | .shapo@leetcode.com     |
+---------+-----------+-------------------------+
Output: 
+---------+-----------+-------------------------+
| user_id | name      | mail                    |
+---------+-----------+-------------------------+
| 1       | Winston   | winston@leetcode.com    |
| 3       | Annabelle | bella-@leetcode.com     |
| 4       | Sally     | sally.come@leetcode.com |
+---------+-----------+-------------------------+
Explanation: 
The mail of user 2 does not have a domain.
The mail of user 5 has the # sign which is not allowed.
The mail of user 6 does not have the leetcode domain.
The mail of user 7 starts with a period.
```

```python
import pandas as pd

# Sample data
data = {
    'user_id': [1, 2, 3, 4, 5, 6, 7],
    'name': ['Winston', 'Jonathan', 'Annabelle', 'Sally', 'Marwan', 'David', 'Shapiro'],
    'mail': ['winston@leetcode.com', 'jonathanisgreat', 'bella-@leetcode.com',
             'sally.come@leetcode.com', 'quarz#2020@leetcode.com',
             'david69@gmail.com', '.shapo@leetcode.com']
}

# Create a DataFrame
users_df = pd.DataFrame(data)

```

```python
import re
import pandas as pd

def is_valid_mail(mail):
    result = re.match(r'^[a-zA-Z][a-zA-Z\d_.-]*@leetcode\.com',mail) is not None
    return result

def valid_emails(users: pd.DataFrame) -> pd.DataFrame:
    filtered_users_df = users[users['mail'].apply(is_valid_mail)]
    return filtered_users_df

# Call the function and display the filtered DataFrame
valid_emails(users_df)

```

```python
The expression re.match(r'^[a-zA-Z][a-zA-Z\d_.-]*@leetcode\.com', mail) 
is a Python code snippet that uses the re module to perform regular expression matching on the input mail string.
This regular expression is designed to check if the given email address belongs to the "leetcode.com" 
domain and follows a specific pattern.

Let's break down the regular expression step by step:

^: This symbol signifies the beginning of the string.

[a-zA-Z]: This part matches the first character of the email address. It specifies that the character 
    must be a lowercase or uppercase letter.

[a-zA-Z\d_.-]*: This section matches the rest of the characters in the username part of the email.
    It allows for a sequence of characters that can be letters (both uppercase and lowercase), digits, underscores,
    periods, or hyphens. The * at the end indicates that this sequence can repeat zero or more times.

@leetcode\.com: This part matches the domain part of the email address. It checks for the exact string "@leetcode.com".

The entire regular expression checks the following pattern:

The email address must start with a letter (lowercase or uppercase).
The email address can then be followed by a sequence of letters, digits, underscores, periods, or hyphens.
The email address must end with "@leetcode.com".
The re.match() function attempts to match the given regular expression pattern at the beginning of the mail string.
If the pattern matches the entire string, the function will return a match object, which is not None. 
If the pattern doesn't match, it will return None.

So, if the expression re.match(r'^[a-zA-Z][a-zA-Z\d_.-]*@leetcode\.com', mail) returns a non-None result,
it means that the mail string is an email address that belongs to the "leetcode.com" domain and follows the specified pattern.






```
